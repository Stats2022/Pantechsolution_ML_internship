# Import library

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Load dataset

```python
data=pd.read_csv('C:\datasets\project7brestcancer.csv')
data
```

# for finding missing value in data and categorical feature

```python
data.isna().any()
data1=data.select_dtypes(exclude='number');data1.head()
```

# convert the attribute data into integer by maping or LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder
Lb=LabelEncoder()
data['diagnosis']=Lb.fit_transform(data.diagnosis)
```

```python
data
```

# Drop the independent feature related to output from the data

```python
data=data.drop(['id'],axis=1);data
```

# Separate the dataset into dependent and indepedent parts

```python
x=data.drop(['diagnosis'],axis=1)
y=data['diagnosis']
```

# Spliting the datasets into train and test data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

# feature scaling 

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

# fitting the models to the above dataset

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import *
```

```python
models=[]
models.append(('DC',DecisionTreeClassifier(random_state=0)))
models.append(('RD',RandomForestClassifier(random_state=0)))
models.append(('SV',SVC()))
models.append(('NB',GaussianNB()))
models.append(('Knn',KNeighborsClassifier()))
models.append(('LR',LogisticRegression()))
```

```python
models
```

```python
accuracy=[]
names=[]
for name,model in models:
               model.fit(x_train,y_train)
               ypred=model.predict(x_test)
               accuracy.append(accuracy_score(y_test,ypred))
               names.append(name)
               print('accuracy of',name,'model=%.3f'%(accuracy_score(y_test,ypred)*100),'%')
```

# Evaluation of model through ROC_Score

```python
from sklearn.model_selection import StratifiedKFold,cross_val_score
results=[]
names=[]
res=[]
for name,model in models:
    st=StratifiedKFold(n_splits=10)
    cvs=cross_val_score(model,x_train,y_train,cv=st,scoring='accuracy')
    names.append(name)
    results.append(cvs)
    res.append(cvs.mean())
    print('%s:%f(%f)'%(name,cvs.mean(),cvs.std()))
```

# graphical representation

```python
import matplotlib.pyplot as plt
plt.ylim(0.90,1)
plt.title('models comparisons')
plt.xlabel('model names')
plt.ylabel('cross_val_score')
plt.bar(names,res,color='red')
plt.show()
```

# From the graph we can say that LogisticRegression model give the best prediction of having the breast cancer or not with approximately 98% accuracy.
